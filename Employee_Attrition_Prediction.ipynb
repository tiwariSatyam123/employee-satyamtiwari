```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype=='object':
        df[col] = le.fit_transform(df[col])
```

```python
X = df.drop('Attrition', axis=1)

y = df['Attrition']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
selector = SelectKBest(
    score_func=chi2,
    k=10
)

X_new = selector.fit_transform(X,y)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_res,y_res = smote.fit_resample(X,y)
```

```python
pd.Series(y_res).value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X_res,
    y_res,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X_train,y_train)
```

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()

xgb.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import (
accuracy_score,
precision_score,
recall_score,
f1_score,
roc_auc_score
)
```

```python
accuracy_score()
precision_score()
recall_score()
f1_score()
roc_auc_score()
```

```python
results_list = []

# Metrics for Random Forest (already predicted as y_pred)
rf_accuracy = accuracy_score(y_test, y_pred)
rf_precision = precision_score(y_test, y_pred, average='weighted')
rf_recall = recall_score(y_test, y_pred, average='weighted')
rf_f1 = f1_score(y_test, y_pred, average='weighted')
rf_roc_auc = roc_auc_score(y_test, y_pred, average='weighted', multi_class='ovr')

results_list.append({
    'Model': 'Random Forest',
    'Accuracy': rf_accuracy,
    'Precision': rf_precision,
    'Recall': rf_recall,
    'F1-Score': rf_f1,
    'ROC AUC': rf_roc_auc
})
```

```python
# Metrics for Logistic Regression
y_pred_lr = lr.predict(X_test)

lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_precision = precision_score(y_test, y_pred_lr, average='weighted')
lr_recall = recall_score(y_test, y_pred_lr, average='weighted')
lr_f1 = f1_score(y_test, y_pred_lr, average='weighted')
lr_roc_auc = roc_auc_score(y_test, y_pred_lr, average='weighted', multi_class='ovr')

results_list.append({
    'Model': 'Logistic Regression',
    'Accuracy': lr_accuracy,
    'Precision': lr_precision,
    'Recall': lr_recall,
    'F1-Score': lr_f1,
    'ROC AUC': lr_roc_auc
})
```

```python
# Metrics for XGBoost
y_pred_xgb = xgb.predict(X_test)

xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
xgb_precision = precision_score(y_test, y_pred_xgb, average='weighted')
xgb_recall = recall_score(y_test, y_pred_xgb, average='weighted')
xgb_f1 = f1_score(y_test, y_pred_xgb, average='weighted')
xgb_roc_auc = roc_auc_score(y_test, y_pred_xgb, average='weighted', multi_class='ovr')

results_list.append({
    'Model': 'XGBoost',
    'Accuracy': xgb_accuracy,
    'Precision': xgb_precision,
    'Recall': xgb_recall,
    'F1-Score': xgb_f1,
    'ROC AUC': xgb_roc_auc
})
```

### Model Performance Comparison

```python
comparison_df = pd.DataFrame(results_list)
display(comparison_df.round(4))
```
